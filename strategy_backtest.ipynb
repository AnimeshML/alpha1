## Data resampling and strategy calculations

### Strategy Parameters

```python
#Variables for strategy backtesting
analysis_freq = 'W'
history = 40
capital = 1000000
n_stocks = 4
trade_freq = 4
start_date = pd.to_datetime('2000-01-01',format='%Y-%m-%d')
end_date = pd.to_datetime('2019-11-18',format='%Y-%m-%d')
start_index = int(history / trade_freq) + 1

main = main[start_date:end_date]
main
```

### Signal Calculation

```python
# Weekly and historical prices resampling and momentum calculation
price_weekly = main.resample('%s'%(analysis_freq)).last()
val_mom = cbuslib.momentum(price_weekly,history)
```

### Trading logic and position change logic

```python
# Momentum value every four weeks
val_trade_freq_mom  = val_mom.resample('%d%s'%(trade_freq,analysis_freq)).last()
```

```python
# Ranking Stocks based on momentum - ascending order
ranks = val_trade_freq_mom.rank(1, ascending=False,method='first',na_option='bottom')
```

### Rankings 

```python
#thresholding based on ranks
ranks[ranks <= n_stocks] = 1
ranks[ranks > n_stocks] = 0
```

```python
bactest_start_date = pd.to_datetime('2000-9-17',format='%Y-%m-%d')
ranks[:start_index] = 0
```

```python
# Creating Buy and Sell position matrices for every 4 weeks
buy_pos = ranks
sell_pos = ranks.shift(1)
```

### Initial Capital allocation

```python
#allocation per stock, price_monthly consists per 4 weeks
allocation = capital/n_stocks
price_monthly = main.resample('%d%s'%(trade_freq,analysis_freq)).last()
```

```python
# Buy and Sell prices for stocks with highest momentum
buy_price = buy_pos * price_monthly
sell_price = sell_pos * price_monthly
```

```python
# Empty df to store sell_value
sell_value = pd.DataFrame()
```

### Portfolio churning based on top 4 Stocks

```python
# Calculation of turnover every 4 weeks
i=0
for row in range(start_index,len(buy_price)):
    buy_qty=np.where(buy_price[row:row+1]==0, 0, allocation/buy_price[row:row+1])
    sell_p = sell_price[row+1:row+2]
    sell_exp = (sell_p.multiply(buy_qty))
    sell_value= sell_value.append(sell_exp)
    allocation = np.sum(sell_exp.values)/n_stocks
            
```

```python
# Sell quantity per stock every 4 week
sell_qty_mat = sell_value/(sell_price)
sell_qty_mat.fillna(0,inplace=True)
```

```python
# Buy quantity per stock every 4 week
buy_qty_mat = sell_qty_mat.shift(-1)
```

```python
# Buy turnover value every 4 week
buy_value = buy_qty_mat*buy_price
sell_value = sell_value.reindex(buy_value.index)
```

```python
# Extrapolating buy and sell quantities on daily level
buy_qty_daily = buy_qty_mat.resample('B').last().ffill()
sell_qty_daily = sell_qty_mat.resample('B').last().ffill()
buy_qty_daily.drop(bad_days,inplace=True)
sell_qty_daily.drop(bad_days,inplace=True)
buy_qty_daily = buy_qty_daily[start_date:end_date]
sell_qty_daily = sell_qty_daily[start_date:end_date]


```

## Performance Analytics

```python
# DataFrame to store turnover every four weeks
net = abs( buy_qty_mat - sell_qty_mat)*main.resample('%d%s'%(trade_freq,analysis_freq)).last()
net['turnover']  = net.sum(axis=1)
net['portfolio_value'] = buy_value.sum(axis=1)
net['ratio'] = (net['turnover']/net['portfolio_value']).where(net['portfolio_value']>0)
net['52weekavg'] = net['ratio'].rolling(13).mean()
```

```python
# Daily Dollar value of portfolio
buy_value_daily = buy_qty_daily*main
sell_value_daily = sell_qty_daily*main
```

```python
# Calculating daily dollar value
buy_value_daily['daily_value'] = buy_value_daily.sum(axis=1)
sell_value_daily['daily_value'] = sell_value_daily.sum(axis=1)
```

```python
# Plotting daily dollar value
buy_value_daily.iplot(

    y='daily_value',
    opacity=.8,
    xTitle='date',
    yTitle='value',
    title='Dollar value with time, daily',
    color='Blue')
```

```python
# Weekly portfolio dollar value and plotting
buy_value_weekly = buy_value_daily.resample('W').last()
sell_value_weekly = sell_value_daily.resample('W').last()

buy_value_weekly.iplot(
    y='daily_value',
    opacity=.8,
    xTitle = 'Time',
    yTitle = 'Value',
    title = 'Dollar Value with Time, Weekly',
    color='Blue')
```

```python
# Calculating 52 week rolling returns
buy_value_weekly['rolling_returns']= \
((buy_value_weekly['daily_value']/buy_value_weekly['daily_value'].\
  shift(52))-1).where(buy_value_weekly['daily_value'].shift(52)>0)
```

```python
# Cumulative 52 week rolling returns
buy_value_weekly['rolling_returns_cum']=buy_value_weekly['rolling_returns'].cumsum()
```

```python
# Plotting 52 week rolling returns
buy_value_weekly.iplot(
    y='rolling_returns_cum',
    opacity = .8,
    xTitle='Time',
    yTitle='Cumulative Returns',
    title='Rolling cumulative returns with Time, 52 Week',
    color='blue')
```

```python
# Distribution of 52 week rolling returns
buy_value_weekly[[ 'rolling_returns']].iplot(
    kind='hist',
    histnorm='percent',
    barmode='overlay',
    xTitle='Weekly Returns',
    yTitle='Frequency',
    title='52 Week rolling Returns Distribution')
```

```python
# Bar graph of 52 week rolling returns
buy_value_weekly[['rolling_returns']].iplot(
    kind='bar',
    xTitle='Time',
    yTitle='Rolling 52 Week Return',
    title='Rolling 52 week returns',
    color='Blue')
```

```python
# Average portfolio turnover 52 week
net.iplot(y='52weekavg'
          ,kind='bar',
          opacity = .8,
          xTitle = 'Time',
          yTitle = 'Turnover (%)',
          title = 'Average 52 week portfolio turnover',
          filename='turnover.html',
          color='green')
```

```python
# Calculation for maximum 52 week loss 
buy_value_weekly['maxloss']=(buy_value_weekly['daily_value']/\
                             buy_value_weekly['daily_value'].\
                             rolling(52).max()) -1 
```

```python
# Plot for 52 week maximum loss
buy_value_weekly.iplot(
    y='maxloss',
    kind='bar',
    xTitle='Time',
    yTitle='52 Week Loss (%)',
    title='52 Week Loss(%)',
    color='red'
)
```
